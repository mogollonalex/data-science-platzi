```
df1 = pd.DataFrame(np.random.rand(3,2)*10,columns=['a','b'])
df1
```

```
df2 = pd.DataFrame(np.random.rand(3,2)*-1,columns=['a','b'], index = [2,3,4])
df2
```

```
pd.concat([df1,df2])
```

```
pd.concat([df1,df2], axis=1)
```

```
pd.concat([df1,df2], axis=1, join='inner')
```
