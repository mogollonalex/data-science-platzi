```python
df = fundamentals[['ticker', 'calendardate', 'pe', 'payoutratio', 'roa', 'roe']][fundamentals['calendardate']=='2018-12-31']
df
```

```python
import statsmodels.api as sm
```

```python
y = df['pe']
```

```python
y.shape
```

```python
X = df[['payoutratio', 'roa', 'roe']]
```

```python
X.shape
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X)
results = model.fit()
```

```python
results.summary()
```
